```python
class Solution:
    # @param A : list of integers
    # @return a list of list of integers
    def permute(self, A):
        res = []
        def permutations(head, acc, tail=[]):
            if len(head) == 0: 
                acc.append(list(tail))
            else:
                for i in range(len(head)):
                    #print head[0:i],head[i+1:],'\t\t', tail+[head[i]]
                    permutations(head[0:i] + head[i+1:], acc, tail+[head[i]])
        permutations(A, res)
        return res
s = Solution()
A = [0,1,2,5,4,3]
s.permute(A)
```

```python
class Solution:
    # @param x : integer
    # @param n : integer
    # @param d : integer
    # @return an integer
    def pow(self, x, n, d):
        if x==0:
            return 0
        if n==0:
            return 1
        
        x %= d
        res = 1
        while n>0:
            if n%2:
                res = (res*x)%d
            x = (x*x)%d
            n = n>>1
        
        return res
        
        
        
        
        
s = Solution()
x, n, d = 2, 30, 10
x, n, d = 71045970, 41535484, 64735492 # 20805472

# 30 = 16+8+4+2 = 11110
# 22 = 16+4+2 = 10110

# 2^1 = 2
# 2^2 = 4
# 2^4 = 6
# 2^8 = 6
# 2^16= 6
# 2^32= 6
s.pow(x,n,d)
```

```python
class Solution:
    # @param x : integer
    # @param n : integer
    # @param d : integer
    # @return an integer
    def pow(self, x, n, d):
        # return x^n % d
        init_x = x
        init_resid = x%d
        x = x%d
        
        cycle = 1
        isCycle = False
        
        for i in range(n-1):
            x = x*init_resid
            #print 'xxxx', x
            x = x%d
            #print 'x' , x
            if x==init_resid:
                isCycle = True
                break
            cycle += 1
            
                
        if isCycle:
            #print 'cycleeee', cycle
            enough_iter = n%cycle -1
            #print 'enough_iter', enough_iter, '\n\tn%cycle',n,'%', cycle
            for i in range(enough_iter):
                init_x = (init_x*init_resid)%d
            return init_x%d
        else:
            return x

        
s = Solution()
x, n, d = 2, 4, 10

x, n, d = 71045970, 41535484, 64735492

# 2 4 8 16 32 64 128 256
# 2 4 8  6  2  4   8   6
# 2 4 3  1  2  4
s.pow(x,n,d)
```

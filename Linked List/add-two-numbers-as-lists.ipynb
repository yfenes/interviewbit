```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    # @param A : head node of linked list
    # @param B : head node of linked list
    # @return the head node in the linked list
    def addTwoNumbers(self, A, B):
        
        # B is always bigger
        if self.length(A)>self.length(B):
            A,B = B,A
        
        carry = 0
        temp_sum = A.val + B.val + carry
        res = ListNode(temp_sum%10)
        carry = int(temp_sum/10)
        to_return = res
        A = A.next
        B = B.next
        
        while A:
            temp_sum = A.val + B.val + carry
            res.next = ListNode(temp_sum%10)
            carry = int(temp_sum/10)
            res = res.next
            A = A.next
            B = B.next
        
        if B:
            while carry and B:
                temp_sum = B.val + carry
                res.next = ListNode(temp_sum%10)
                carry = int(temp_sum/10)
                res = res.next
                B = B.next
            if carry:
                res.next = ListNode(1)
            else:
                res.next = B
        else:
            if carry:
                res.next = ListNode(1)
        
        return to_return
            
    def length(self,A):
        counter = 0
        while A:
            counter+=1
            A = A.next
        return counter
        
# 342 + 465 = 807
A0 = ListNode(2)
A1 = ListNode(4)
A2 = ListNode(3)

B0 = ListNode(5)
B1 = ListNode(6)
B2 = ListNode(4)

A0.next = A1
A1.next = A2
B0.next = B1
B1.next = B2

AA0 = ListNode(1)
BB0 = ListNode(9)
BB1 = ListNode(9)
BB2 = ListNode(9)

BB0.next = BB1
BB1.next = BB2

s = Solution()
res = s.addTwoNumbers(AA0,BB0)
while res:
    print(res.val)
    res = res.next
```
